```python
# combine all silos


# Specify the directory containing the silo CSV files
silo_directory = 'datasets/nids'

# Initialize an empty DataFrame to store the combined data
combined_data = pd.DataFrame()

# Loop through each silo file
for filename in os.listdir(silo_directory):
    if filename.endswith('.csv'):
        file_path = os.path.join(silo_directory, filename)
        # Read the CSV file
        data = pd.read_csv(file_path)
        # Append the data to the combined_data DataFrame
        combined_data = combined_data.append(data, ignore_index=True)

# Save the combined data to a new CSV file
combined_data.to_csv('datasets/nids/combined_data.csv', index=False)
```

```python
def get_subset_by_ratio(df, label_column, ratio):
    # Group the DataFrame by the label column
    grouped = df.groupby(label_column)

    # Initialize an empty DataFrame to store the subset
    subset = pd.DataFrame()

    # Iterate over the groups
    for label, group in grouped:
        # Determine the number of samples to keep for this class
        num_samples = int(len(group) * ratio)
        
        # Sample the required number of samples from this class
        sampled_group = group.sample(n=num_samples, random_state=42)
        
        # Append the sampled group to the subset DataFrame
        subset = subset.append(sampled_group)

    # Reset the index of the subset DataFrame
    subset = subset.reset_index(drop=True)
    
    return subset
```

```python
def create_equal_proportion_subset(dataframe, label_column):
    # Count the occurrences of each attack type
    attack_counts = dataframe[label_column].value_counts()

    # Find the minimum count among all attack types
    min_count = min(attack_counts)

    # Create an empty DataFrame to store the subset
    subset = pd.DataFrame()

    # Iterate over each attack type
    for attack_type in attack_counts.index:
        # Select a subset with the minimum count for the current attack type
        temp_subset = dataframe[dataframe[label_column] == attack_type].sample(n=round(min_count/50), random_state=42)

        # Append the subset to the main subset DataFrame
        subset = pd.concat([subset, temp_subset])

    # Shuffle the subset DataFrame
    subset = subset.sample(frac=1, random_state=42)

    return subset
```
